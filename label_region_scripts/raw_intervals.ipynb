```python
def get_EncodeID_tfname(filename):
    encode_id = filename.split('/')[-1]
    for path,folders,files in os.walk(filename):
        if 'metadata.org.json' in files:
            path_to_metadata = os.path.join(path,'metadata.org.json')
            with open(path_to_metadata,'r') as f:
                metadata = json.load(f)
                tf_name = metadata['target']['label']
                return(encode_id,tf_name)
    
```

```python
def get_bed_file(filename):
    def chk_rep1_rep2_idr(filename):
        for path,folders,files in os.walk(filename):
            if 'optimal_idr' in path and 'rep1_rep2' in path:
                if len(files)>0:
                    path = os.path.join(path,files[0])
                    return path
    def chk_rep3_rep4_idr(filename):
        for path,folders,files in os.walk(filename):
            if 'optimal_idr' in path and 'rep3_rep4' in path:
                if len(files)>0:
                    path = os.path.join(path,files[0])
                    return path            
    def chk_replicated_peaks(filename):
        for path,folders,files in os.walk(filename):
            if 'replicated_peaks' in path and 'rep1_rep2' in path:
                if len(files)>0:
                    path = os.path.join(path,files[0])
                    return path
        
    def chk_idr(filename):
        for path,folders,files in os.walk(filename):
            if 'idr' in path and len(files)>0:
                if 'bed.gz' in files[0]:
                    path = os.path.join(path,files[0])
                    return(path)
   
    def chk_any(filename):
        for path,folders,files in os.walk(filename):
            if len(files)>0:
                if 'bed.gz' in files[0]:
                    path = os.path.join(path,files[0])
                    return(path)
   
    if chk_rep1_rep2_idr(filename):
        return(chk_rep1_rep2_idr(filename))
    if chk_rep3_rep4_idr(filename):
        return(chk_rep3_rep4_idr(filename))
    if chk_replicated_peaks(filename):
        return(chk_replicated_peaks(filename))
    elif chk_idr(filename):
        return(chk_idr(filename))
    elif chk_any(filename):
        return chk_any(filename)
    
```

```python
def create_raw_intervals_json(TF,path_to_narrowpeak,raw_file_name,cell = 'TAIR10',
                              region_bed ='/srv/scratch/manyu/genome_files/TAIR10.chrom.bed' ):
    vals_dict = {}
    vals_dict = {cell: {'feature_beds': {TF: path_to_narrowpeak},
  'region_bed': region_bed},
 'task_names': [TF]}  
    with open(raw_file_name,'w') as f:
        json.dump(vals_dict,f)
    print("Created raw intervals file for {} at {}".format(TF,raw_file_name))    
```

```python
for filename in file_list:
    try:
        base_path = '/oak/stanford/groups/akundaje/manyu/train_all_models/hg38/train_K562/label_regions/raw_intervals/'
        encode_id,TF = get_EncodeID_tfname(filename)
        bedfile = get_bed_file(filename)
        cell = 'K562'
        region_bed = path_to_raw_intervals_bed
        raw_file_name = base_path+encode_id+'_'+TF+'_'+'raw_intervals.json'
        create_raw_intervals_json(TF,bedfile,raw_file_name,cell,region_bed)
        print("Processed {}".format(encode_id))
    except Exception as e:
        print(e)
        pass
```
